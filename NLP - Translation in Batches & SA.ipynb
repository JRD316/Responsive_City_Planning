```python
import pandas as pd
from googletrans import Translator
```

```python
from tqdm import tqdm
```

```python
translator = Translator()
```

```python
# 3) Robust batch translator that falls back to single calls on error
def translate_in_batches(texts, src='es', dest='en', batch_size=100):
    translations = []
    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i : i + batch_size]
        try:
            results = translator.translate(batch, src=src, dest=dest)
            if not isinstance(results, list):
                results = [results]
            translations.extend([r.text for r in results])
        except Exception:
            for t in batch:
                if not t.strip():
                    translations.append('')
                else:
                    try:
                        r = translator.translate(t, src=src, dest=dest)
                        translations.append(r.text)
                    except:
                        translations.append('')
    return translations
```

```python
df = pd.read_excel('socialmedia.xlsx')
```

```python
texts = df['content'].fillna('').astype(str).tolist()
df['content_en'] = translate_in_batches(texts, batch_size=100)
```

```python
# Preview the first few rows of original and translated columns
print(df[['content', 'content_en']].head())
```

```python
df['content_en'] = df.pop('content_en')
```

```python
df.to_excel('socialmedia_translated.xlsx', index=False)
```

```python
df = pd.read_excel('socialmedia_translated.xlsx')
```

```python
print(df.head)
```

```python

# 4. Load the English-only sentiment model
model = tweetnlp.Sentiment()

# 5. Define a helper to analyze English sentiment
def analyze_sentiment(text):
    if not isinstance(text, str) or not text.strip():
        return pd.Series({
            'sentiment':   None,
            'p_negative':  0.0,
            'p_neutral':   0.0,
            'p_positive':  0.0,
        })
    out = model.sentiment(text, return_probability=True)
    probs = out['probability']
    return pd.Series({
        'sentiment':   out['label'],
        'p_negative':  probs.get('negative', 0.0),
        'p_neutral':   probs.get('neutral',  0.0),
        'p_positive':  probs.get('positive', 0.0),
    })

# 6. Apply the helper to your English column
sentiment_results = df['content_en'].apply(analyze_sentiment)

# 7. Combine results into a new DataFrame
df_out = pd.concat([df, sentiment_results], axis=1)

# 8. Save to a new Excel file
df_out.to_excel('socialmedia_translated_sentiment.xlsx', index=False)
```
