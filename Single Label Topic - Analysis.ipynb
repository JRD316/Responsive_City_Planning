```python
from tqdm import tqdm
import pandas as pd
import tweetnlp
```

```python
df = pd.read_excel('socialmedia_translated_sentiment.xlsx')
```

```python
print(df.head)
```

```python
model = tweetnlp.load_model('topic_classification', multi_label=False)
```

```python
def classify_topic(text):
    if not isinstance(text, str) or not text.strip():
        return pd.Series({'topic': None, 'topic_prob': 0.0})
    out = model.topic(text, return_probability=True)
    label = out['label']
    prob  = out['probability'].get(label, 0.0)
    return pd.Series({'topic': label, 'topic_prob': prob})
```

```python
topic_results = df['content_en'].apply(classify_topic)
```

```python
df_out = pd.concat([df, topic_results], axis=1)
df_out.to_excel('socialmedia_topic_singlelabel.xlsx', index=False)
```
